# Camera TTC Remarks

Among the combination of the Detectors and Descriptors used below, most of them produced a normal trend towards a smaller TTC, which was expected, as the front car in the scene approximates the EGO car.

However, some of them produced more erratic behaviour as others:
* By far, the worst detector / descriptor was ORB. Besides producing unrealistic very high TTC, there were frames that it didn't manage to either identify or produce descriptions for the matchers to match the points confidently. 
* Brisk Detector / Descriptor produced spikes in the TTC from frame to frame, in a scale which I believe would be prohibitive (> 10 seconds difference)
* **Sift Detector + Brief Descriptor** and **Fast Detector + Freak Descriptor** produced one spike in the analyzed frames, but further investigation would be needed to identify the possible reasons.

* **All the other combinations of detectors / descriptors below could be used more confidently based on the image scenes used in the project**


```python
df = pd.read_csv("dat/analysis.csv", sep=";")
df.head()
```

```python
plt.figure(figsize=[16,4])
for keys, df_group in df.groupby(["detectorType", "descriptorType"]):
    plt.plot(df_group["imageIndex"], df_group["cameraTTS"], label=keys)
plt.xlabel("imageIndex")
plt.ylabel("cameraTTS(sec)")
plt.title("Comparison of TTS estimation by different Detectors/Descriptors combinations")
plt.legend()
plt.ylim([0,37])
plt.grid()

custom_tick_labels = np.arange(0,19,1)
plt.xticks(custom_tick_labels,custom_tick_labels);
```
